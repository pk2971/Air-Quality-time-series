```python
import time
import random
def get_data(link):
 return requests.get(link)
 
def get_soup(data):
   soup = bs(data.content)
   return soup
 
def get_parser(data):
   text = data.text
   htmlParse = BeautifulSoup(text, "html.parser")
   return htmlParse

def construct_data_list(links):
  # looping through the links
  link_list = links

  # creating a new list            
  hansard_info_list = []
  n=0

  # for every link in the list of links
  for link in link_list:


    # appending the string to the link
    # to get the entire link
    if(n<10):
      #print("Extracting ",link)
      #print(n)
      string = "https://api.parliament.uk"
      link = string + link

    # creating a dictionary entry called hansard_info
      hansard_info = {}
      data = get_data(link)
      soup = get_soup(data)

    # getting the date for each link
      date=soup.find(class_="sitting-day").get_text(" ", strip = True)
      if date is not None:
        result=date
      else:
        result='None'
      hansard_info['date'] = result
      parser = get_parser(data)
      title=parser.select('h1.title')[0].text.strip()
      if title is not None:
        result=title
      else:
        result='None'
    # extracting the title for each link
      hansard_info['title'] = title
    # extracting the conversation from the webpage for each link
      conversation=[p.get_text(strip=True, separator='\n') for p in soup.find_all("p")]
      if conversation is not None:
        result=conversation
      else:
        result='None'
      hansard_info['conversation'] = result
    # getting rid of "noticed a typo?"
      hansard_info['conversation'] = hansard_info['conversation'][:-1]
    # appending this dictionary entry the list of ditionary entries
      hansard_info_list.append(hansard_info)
      n+=1
    else:
      x=random.randrange(1,4)
      print("Sleeping for ",x," seconds")
      time.sleep(x)
      n=0
  return hansard_info_list
```

Convert list of dictionary entries into a csv file

```python
#run from here
import pandas as pd
```

```python
# converting a list of dictionary elements into a csv file
import csv
def importToCSV1(links_small):
  to_csv = construct_data_list(links_small)

  keys = to_csv[0].keys()

  with open('hansard1.csv', 'w', newline='') as output_file:
      dict_writer = csv.DictWriter(output_file, keys)
      dict_writer.writeheader()
      dict_writer.writerows(to_csv)
def importToCSV2(links_small):
  to_csv = construct_data_list(links_small)

  keys = to_csv[0].keys()

  with open('hansard2.csv', 'w', newline='') as output_file:
      dict_writer = csv.DictWriter(output_file, keys)
      dict_writer.writeheader()
      dict_writer.writerows(to_csv)
```
